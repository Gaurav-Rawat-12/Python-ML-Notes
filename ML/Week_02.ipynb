## Simple Linear Regression (SLR) — Practical Notes (with code)

---

## 1. What it is (Real-world view)

* Used to predict a **continuous value** from **one feature**
* Example:

  * Experience → Salary
  * Ad spend → Sales
  * House size → Price

👉 Model learns a straight line:
[
y = mx + b
]

* **m (slope)** → how much y changes per unit x
* **b (intercept)** → value of y when x = 0

---

## 2. Intuition (very important)

* Model tries to draw a line that is **closest to all points**
* “Closeness” = minimize error
* Error = difference between actual and predicted

👉 This is called **Ordinary Least Squares (OLS)**

---

## 3. Basic Workflow (Industry style)

1. Load data
2. Separate X and y
3. Train model
4. Predict
5. Evaluate

---

## 4. Code (Clean + Minimal)

### Step 1: Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

---

### Step 2: Sample Data

```python
# Example: Experience vs Salary
data = pd.DataFrame({
    'experience': [1, 2, 3, 4, 5],
    'salary': [30000, 35000, 50000, 60000, 70000]
})
```

---

### Step 3: Split X and y

```python
X = data[['experience']]   # 2D
y = data['salary']         # 1D
```

---

### Step 4: Train Model

```python
model = LinearRegression()
model.fit(X, y)
```

---

### Step 5: Prediction

```python
pred = model.predict([[6]])
print(pred)
```

---

### Step 6: Model Parameters

```python
print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```

---

### Step 7: Visualization (important)

```python
plt.scatter(X, y)                  # actual points
plt.plot(X, model.predict(X))      # regression line
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

---

## 5. Evaluation (Industry must)

### R² Score

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)
print(r2_score(y, y_pred))
```

* **R² = 1** → perfect fit
* **R² ≈ 0** → useless model

---

## 6. Common Mistakes (very important)

❌ Passing 1D X

```python
X = data['experience']   # wrong
```

✔ Correct:

```python
X = data[['experience']]
```

---

❌ Predict without 2D input

```python
model.predict([6])  # wrong
```

✔ Correct:

```python
model.predict([[6]])
```

---

❌ Over-trusting model

* Linear relation must exist
* Otherwise → bad predictions

---

## 7. When to use (Industry)

Use SLR when:

* Relationship looks roughly straight
* Only **one strong feature** matters
* Need quick baseline model

---

## 8. When NOT to use

* Multiple features → use Multiple Linear Regression
* Non-linear pattern → use polynomial / tree models

---

## 9. Viva / Interview Quick Answers

**Q: What is Linear Regression?**
→ A supervised learning algorithm used to predict continuous values using a linear relationship.

**Q: What is slope?**
→ Change in output per unit change in input.

**Q: What is intercept?**
→ Value of output when input is zero.

**Q: What is OLS?**
→ Method to minimize squared errors.

**Q: Why squared error?**
→ Penalizes large errors more.
