## Machine Learning Development Life Cycle (MLDLC)

---

### **1. Problem Definition**

* Define **business goal**
* Convert into **ML problem**

Example:

* Business: “Predict stock price”
* ML: “Time-series regression / forecasting”

---

### **2. Data Collection**

* Sources:

  * APIs (like `yfinance`)
  * Databases
  * CSV / logs
* Ensure:

  * sufficient data
  * relevant features

---

### **3. Data Preprocessing**

* Handle missing values
* Remove duplicates
* Normalize / scale (e.g., MinMaxScaler)
* Encode categorical data

Goal: **clean + usable dataset**

---

### **4. Exploratory Data Analysis (EDA)**

* Understand patterns:

  * trends
  * correlations
* Tools:

  * Pandas
  * Matplotlib / Seaborn

Example:

* price trends
* volatility

---

### **5. Feature Engineering**

* Create new features:

  * RSI, MACD (in your project)
  * moving averages
* Select important features

This step = **huge impact on performance**

---

### **6. Model Selection**

Choose algorithm based on problem:

* Regression → Linear, LSTM
* Classification → Logistic, SVM
* Time-series → LSTM, ARIMA

---

### **7. Model Training**

* Split data:

  * Train / Validation / Test
* Train model on training set
* Tune hyperparameters

---

### **8. Model Evaluation**

Metrics depend on problem:

* Regression:

  * RMSE
  * MAE

Evaluate on **unseen test data**

---

### **9. Model Optimization**

* Improve performance:

  * Hyperparameter tuning
  * Regularization
  * Feature selection

---

### **10. Deployment**

* Convert model into usable system:

  * API (FastAPI)
  * Web app (React)

---

### **11. Monitoring & Maintenance**

* Track performance over time
* Retrain with new data
* Handle data drift

---

### **Full Flow (Simple)**

```text
Problem → Data → Clean → Analyze → Features → Model → Train → Evaluate → Deploy → Monitor
```

---

### **Real-World Insight**

* 70–80% work = **data + preprocessing**
* Model = only small part
* Deployment = what companies care about
