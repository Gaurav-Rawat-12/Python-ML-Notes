**Data Accessing:**

Data accessing means fetching data from the dataset or, in our case, from the list. Fetching data from a list is easy if you already know the index. If not, we use a loop.

But in this case, we have multiple dictionaries inside the list—so how do we loop through this? This is something that is often not explained clearly in many tutorials.

Let’s suppose we want to access (fetch) the marks of Sonia from the list. This process is also called traversing data or accessing values.

We know that a dictionary stores data in key-value pairs. Since we already know the value ("Sonia"), we can loop through the list and check the `"name"` key to fetch Sonia’s marks.


```python
for x in students: # Iterating/Looping through the list of students
    if x["name"] == "Sonia": # Checking if the name of the student is "Sonia"
        print(x["marks"])
```

As you can see, we wrote x["name"] and not simply name or "name" or [name]. Just like [] is used for lists and " " is used for strings, [" "] is used for keys in a dictionary.

Extract: Now we know how to access the data, let's discuss how to extract data. Extraction is collecting a certain type of data from a dataset. For example: extracting marks from students.

Let's see the code for the example. We will extract all the marks from the students and put them in a new list called marks_list.

```python
marks_list = [] # Creating an empty list to store the marks of all students

for x in students: 
    marks_list.append(x["marks"]) # Appending the marks of each student to marks_list
print(marks_list)   

### Other examples ###

# Extract all prices
# prices.append(product["price"])

# Extract features
# features.append(user["age"]) 
```

Now, in the industry, we check everything. A small mistake can lead to a bigger problem. That's why we should make sure the code keeps running without errors, even if a key (here, `"marks"`) doesn’t exist. To do this, we will simply add an if condition.

```python
marks_list = []

for x in students:
    if "marks" in x: # Checking if the key "marks" exists in the dictionary x
        marks_list.append(x["marks"])
print(marks_list)
```

```python
# Small and cleaner version of the code above using list comprehension 
marks_list = [x["marks"] for x in students if "marks" in x]
```

```python
avg = sum(marks_list) / len(marks_list)
```

The above code is used to calculate the average of marks. But just as I told you earlier, in the industry we write code assuming it can break. Example problems with the above code:

Empty list
Missing values
Wrong types

So we will make the code a little more stable like this:

```python
if len(marks_list) > 0:
    avg = sum(marks_list) / len(marks_list)
else:
    avg = 0

# Cleaner way to write the above code
avg = sum(marks_list) / len(marks_list) if marks_list else 0
```

As you can see, the above code only handles the empty list and not missing values or wrong types. We can handle them by using the `isinstance` function.

```python
clean_marks = []

for x in marks_list:
    if isinstance(x, (int, float)):
        clean_marks.append(x)

avg = sum(clean_marks) / len(clean_marks) if clean_marks else 0
```

Now let's combine all **CRUD operations** (Add, Read, Update, Delete) and **basic analytics** (average, topper, pass/fail) into a single CLI-based Student Manager system.

```python
# Note: This is a simple CLI project for learning purposes, not production-level code

import copy

student = copy.deepcopy(students_01)


# ---------- ADD ----------
def add_student(student, name, marks):
    student.append({"name": name, "marks": marks})

# ---------- REMOVE ----------
def remove_student(student, name):
    student[:] = [x for x in student if x["name"] != name]

# ---------- SEARCH ----------
def search_student(student, name):
    for x in student:
        if x["name"] == name:
            return x
    return None

# ---------- SHOW ----------
def show_students(student):
    for x in student:
        print(x)

# ---------- UPDATE ----------
def update_marks(student, name, new_marks):
    for x in student:
        if x["name"] == name:
            x["marks"] = new_marks

# ---------- TOPPER ----------
def get_topper(student):
    clean = [x for x in student if "marks" in x and isinstance(x["marks"], (int, float))]
    return max(clean, key=lambda x: x["marks"]) if clean else None

# ---------- AVERAGE ----------
def get_average(student):
    marks = [x["marks"] for x in student if isinstance(x.get("marks"), (int, float))]
    return sum(marks)/len(marks) if marks else 0

# ---------- PASS/FAIL ----------
def count_pass_fail(student, pass_marks=40):
    passed = sum(1 for x in student if x.get("marks", 0) >= pass_marks)
    failed = len(student) - passed
    return passed, failed


# ---------- CLI ----------
while True:
    print("\n1.Add  2.Remove  3.Search  4.Show  5.Update  6.Topper  7.Average  8.Pass/Fail  9.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Name: ")
        marks = float(input("Marks: "))
        add_student(student, name, marks)

    elif choice == "2":
        name = input("Name: ")
        remove_student(student, name)

    elif choice == "3":
        name = input("Name: ")
        print(search_student(student, name))

    elif choice == "4":
        show_students(student)

    elif choice == "5":
        name = input("Name: ")
        marks = float(input("New Marks: "))
        update_marks(student, name, marks)

    elif choice == "6":
        top = get_topper(student)
        print(top if top else "No data")

    elif choice == "7":
        print("Average:", get_average(student))

    elif choice == "8":
        p, f = count_pass_fail(student)
        print("Passed:", p, "Failed:", f)

    elif choice == "9":
        break

    else:
        print("Invalid choice")
```

# Error Handling

Now the above code looks good and all, but there are many little hidden errors in it. Like logical errors, value errors, etc. As a professional, it's important to write code while keeping errors in mind.

I'm not going to write anything in detail because this is something that comes from experience.

Of course, you don't have to remember all types of errors, but you must know the main ones.

There are basically two main ways to handle errors: the `try-except` method and one special function.

👉 **try-except** = handles runtime errors (exceptions)

👉 **.get()** = prevents a specific error (`KeyError`) before it happens

---

### 1. ValueError

👉 Wrong type conversion

```python

int("abc")  # ❌


#✔ Handle:

try:
    x = int(input())
except ValueError:
    print("Invalid number")
```

### 2. TypeError

👉 Wrong data type usage


```python
"2" + 2  # ❌


#✔ Handle:

try:
    result = "2" + 2
except TypeError:
    print("Type mismatch")
```

### 3. KeyError

👉 Missing dictionary key

```python
student["marks"]  # ❌ if not present


# ✔ Handle (2 ways):

# 👉 Better way (avoid error):

student.get("marks", 0)

# 👉 Or:

try:
    marks = student["marks"]
except KeyError:
    print("Key missing")
```

### 4. IndexError

👉 List index out of range

```python
lst = [1, 2]
print(lst[5])  # ❌


# ✔ Handle:

try:
    print(lst[5])
except IndexError:
    print("Index out of range")
```


### 5. ZeroDivisionError

👉 Division by zero

```python
10 / 0  # ❌


# ✔ Handle:

try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

### 6. AttributeError

👉 Object doesn’t have that method

```python
x = 10
x.append(5)  # ❌
```

### Pattern You Must Remember
Most programs follow this:

```python
try:
    # risky code
except SpecificError:
    # handle it
```

- Never blindly use except:

- Always try to use **specific error**

Most of the time we use try-except method with methods likes if-else, while, etc. 

**For example:**

```python
while True:
    try:
        x = int(input())

        if ______:
            break
        else:
            print("Invalid")

    except ______:
        print("Error")
```

### Real Example

```python
while True:
    try:
        x = int(input())

        if x > 0:
            break
        else:
            print("Invalid")

    except ValueError:
        print("Wrong input")
```
