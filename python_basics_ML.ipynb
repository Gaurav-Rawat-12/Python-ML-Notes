**Data Accessing:** Data accessing is fetching the data from the dataset or in our case from the list. Now fetching the data from the list is very easy if you already know the index, if not we use loop. 

But in this case we have multiple dictionaries inside the list, how are we going to loop through this? This was never taught in online videos, right? 

First let's suppose we need to access(fetch) the marks of Sonia form the list. This is also called Traverse data or Access values(Fancy terms). So, we know that the dict store the data in key : Value pair. We already know the value(Sonia), so we will simply loop through Key called name to fetch the marks of Sonia. 

```python
for x in students: # Iterating/Looping through the list of students
    if x["name"] == "Sonia": # Checking if the name of the student is "Sonia"
        print(x["marks"])
```

As you can see we wrote x["name"] and not simply name or "name" or [name]. Just like [] is used list and " " is used for string, [" "] is used for key of dict. 

**Extract**: Now we know how to access the data let's discuss how to extract data. Exctraction is collecting a certain type of data from the dataset. For eg: Extracting marks form students. 

Let's see the code for the example, we will extract all the marks form the students and will put in new list called marks_list

```python
marks_list = [] # Creating an empty list to store the marks of all students

for x in students: 
    marks_list.append(x["marks"]) # Appending the marks of each student to the marks_list
print(marks_list)   

### Other examples ###

''' Extract all prices
prices.append(product["price"])

# Extract features
features.append(user["age"]) '''
```

Now, in industry we check everything, a small mistake can lead to a bigger problem. That's why we should make sure the code should keep running without error even if key (here,"marks") didn't exists. To do this we will simply add a if funciton. 

```python
marks_list = []

for x in students:
    if "marks" in x: # Checking if the key "marks" exists in the dictionary x
        marks_list.append(x["marks"])
print(marks_list)
```

```python
# Small and cleaner version of the code above using list comprehension 
marks_list = [x["marks"] for x in students if "marks" in x]
```

The above code is the code to calculate the avg of marks. But just as I told you eariler in industry we write the code assuming it will or can break. Example problems for above code:
- Empty list 
- Missing values 
- Wrong types 

So we will make the code a little more steady like this:

```python
if len(marks_list) > 0:
    avg = sum(marks_list) / len(marks_list)
else:
    avg = 0

# More cleaner way to write the above code
avg = sum(marks_list) / len(marks_list) if marks_list else 0
```

As you can see the above code only handles the Empty list and not Missing values or wrong types. We can handle them by using isinstance function.

```python
clean_marks = []

for x in marks_list:
    if isinstance(x, (int, float)):
        clean_marks.append(x)

avg = sum(clean_marks) / len(clean_marks) if clean_marks else 0
```

**Now let's combine all CRUD operations (Add, Read, Update, Delete) and basic analytics (average, topper, pass/fail) into a single CLI-based Student Manager system.**

```python
import copy

student = copy.deepcopy(students_01)


# ---------- ADD ----------
def add_student(student, name, marks):
    student.append({"name": name, "marks": marks})

# ---------- REMOVE ----------
def remove_student(student, name):
    student[:] = [x for x in student if x["name"] != name]

# ---------- SEARCH ----------
def search_student(student, name):
    for x in student:
        if x["name"] == name:
            return x
    return None

# ---------- SHOW ----------
def show_students(student):
    for x in student:
        print(x)

# ---------- UPDATE ----------
def update_marks(student, name, new_marks):
    for x in student:
        if x["name"] == name:
            x["marks"] = new_marks

# ---------- TOPPER ----------
def get_topper(student):
    clean = [x for x in student if "marks" in x and isinstance(x["marks"], (int, float))]
    return max(clean, key=lambda x: x["marks"]) if clean else None

# ---------- AVERAGE ----------
def get_average(student):
    marks = [x["marks"] for x in student if isinstance(x.get("marks"), (int, float))]
    return sum(marks)/len(marks) if marks else 0

# ---------- PASS/FAIL ----------
def count_pass_fail(student, pass_marks=40):
    passed = sum(1 for x in student if x.get("marks", 0) >= pass_marks)
    failed = len(student) - passed
    return passed, failed


# ---------- CLI ----------
while True:
    print("\n1.Add  2.Remove  3.Search  4.Show  5.Update  6.Topper  7.Average  8.Pass/Fail  9.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Name: ")
        marks = float(input("Marks: "))
        add_student(student, name, marks)

    elif choice == "2":
        name = input("Name: ")
        remove_student(student, name)

    elif choice == "3":
        name = input("Name: ")
        print(search_student(student, name))

    elif choice == "4":
        show_students(student)

    elif choice == "5":
        name = input("Name: ")
        marks = float(input("New Marks: "))
        update_marks(student, name, marks)

    elif choice == "6":
        top = get_topper(student)
        print(top if top else "No data")

    elif choice == "7":
        print("Average:", get_average(student))

    elif choice == "8":
        p, f = count_pass_fail(student)
        print("Passed:", p, "Failed:", f)

    elif choice == "9":
        break

    else:
        print("Invalid choice")
```
